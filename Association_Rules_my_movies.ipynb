```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movie_data=pd.read_csv("C:\\excelr\\datasets\\my_movies.csv")
movie_data
```

```python
movie_data.drop(["V1","V2","V3","V4","V5"],inplace=True,axis = 1)
```

```python
movie_data
```

```python
frequent_itemsets = apriori(movie_data,min_support=0.07,use_colnames=True)
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
```

```python
rules[(rules['lift']>= 6) & (rules['confidence'] >= 0.8)]
```
